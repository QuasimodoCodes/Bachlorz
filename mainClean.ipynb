```python
import pandas as pd
import json

# Load the JSON file with utf-8 encoding
with open('Data/structured_evaluation_and_technical_annexes.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Prepare the structure for the DataFrame
columns = [
    'Tender Title', 'Evaluation and Award', 'Annex - Technical/Tender Specifications', 'Tables', 
    'Exclusion Criteria', 'Selection Criteria', 'Award Criteria', 'Documents to Submit', 
    'Tenderer Requirements', 'Project/Contract Details', 'Criteria Description', 
    'Supporting Documents', 'Impact/Performance Indicators', 'Financial Information', 'Project Timeline'
]

# Initialize an empty list to hold the rows
rows = []

# Loop through the data to extract relevant details
for tender_title, tender_data in data.items():
    # Extract necessary fields from the JSON structure
    evaluation_and_award = tender_data.get('Evaluation and Award', '')
    annex_technical_specifications = tender_data.get('Annex - Technical/Tender Specifications', '')
    tables = tender_data.get('Tables', [])
    exclusion_criteria = tender_data.get('Exclusion Criteria', '')
    selection_criteria = tender_data.get('Selection Criteria', '')
    award_criteria = tender_data.get('Award Criteria', '')
    documents_to_submit = ""  # You can add logic to extract this if available
    tenderer_requirements = ""  # You can extract specific criteria if needed
    project_details = ""  # Can be detailed if available in the data
    criteria_description = ""  # If there are specific criteria descriptions
    supporting_documents = ""  # Extract necessary documents here
    impact_performance_indicators = ""  # Can be extracted based on project impact
    financial_information = ""  # Extract financial details if needed
    project_timeline = ""  # Extract timeline if mentioned

    # Add a row for each tender
    rows.append([
        tender_title,
        evaluation_and_award,
        annex_technical_specifications,
        str(tables),  # Convert tables list to string
        exclusion_criteria,
        selection_criteria,
        award_criteria,
        documents_to_submit,
        tenderer_requirements,
        project_details,
        criteria_description,
        supporting_documents,
        impact_performance_indicators,
        financial_information,
        project_timeline
    ])

# Create a DataFrame
df = pd.DataFrame(rows, columns=columns)

# Export to CSV
df.to_csv('cleaned_tender_data.csv', index=False)

# Show the first few rows of the DataFrame
print(df.head())
```

```python
import nltk
from nltk.corpus import stopwords
import re
from bs4 import BeautifulSoup

# Download stopwords data from NLTK
nltk.download('stopwords')

# Function to clean the text and remove stopwords
def clean_text_and_remove_stopwords(text, stop_words):
    # 1. Remove HTML tags (if applicable)
    text = BeautifulSoup(text, "html.parser").get_text()

    # 2. Remove special characters and control characters (except spaces and letters)
    text = re.sub(r'[^\w\s]', '', text)

    # 3. Standardize whitespace by replacing multiple spaces, tabs, and newlines with a single space
    text = re.sub(r'\s+', ' ', text).strip()

    # 4. Convert to lowercase for consistency
    text = text.lower()

    # 5. Remove stop words
    text = ' '.join([word for word in text.split() if word not in stop_words])

    return text

# Define the languages for stopword removal
languages = ['english', 'french', 'spanish', 'german', 'portuguese']
stop_words = {lang: set(stopwords.words(lang)) for lang in languages}

# Combine all stop words from the selected languages into one set
all_stop_words = set()
for lang in languages:
    all_stop_words.update(stop_words[lang])

# Apply the text cleaning and stopword removal function to the relevant columns
for col in df.columns:
    if df[col].dtype == 'object':  # Only apply cleaning to text columns
        df[col] = df[col].apply(lambda x: clean_text_and_remove_stopwords(str(x), all_stop_words))

# Show cleaned data for the first few rows
print(df.head())
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

""" # Download necessary NLTK data
nltk.download('wordnet')
nltk.download('omw-1.4') """

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to apply lemmatization to each word
def lemmatize_text(text):
    # Tokenize the text into words
    words = text.split()
    
    # Lemmatize each word
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    
    # Join the lemmatized words back into a single string
    return ' '.join(lemmatized_words)

# Apply lemmatization to each relevant column in the DataFrame
for col in df_cleaned.columns:
    if df_cleaned[col].dtype == 'object':  # Only apply lemmatization to text columns
        df_cleaned[col] = df_cleaned[col].apply(lambda x: lemmatize_text(str(x)))

# Show cleaned and lemmatized data for the first few rows
df_cleaned.head()
```
